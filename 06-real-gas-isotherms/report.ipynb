# Data Analysis: Isotherms of Real Gases (Sulfur Hexafluoride)

## Project Overview
This project investigates the phase behavior and thermodynamic properties of a real gas ($SF_6$) through the analysis of pressure-volume (p-V) isotherms. The study focuses on modeling experimental data with the van der Waals equation and determining critical parameters through numerical methods.

### Key Areas of Investigation:
1. **Equation of State Modeling**: Evaluating the van der Waals model against experimental isotherms for a real gas.
2. **Critical Point Analysis**: Determining critical pressure ($p_k$), volume ($V_{mk}$), and temperature ($T_k$) through the analysis of inflection points.
3. **Maxwell Construction**: Analyzing phase transitions and liquid-vapor equilibrium regions below the critical temperature.
4. **Parameter Estimation**: Extracting the cohesion pressure ($a$) and co-volume ($b$) parameters from experimental data.

---

# 1. Theoretical Basis

## 1.1 Equation of State
The state of an ideal gas is described by the equation $pV_m = RT$. This model, however, neglects the finite volume of particles and intermolecular forces. For real gases like sulfur hexafluoride ($SF_6$), these factors are significant. The system is described by the **van der Waals equation**:
$$\left(p + \frac{a}{V_m^2}\right)(V_m - b) = RT$$
where:
* $p$ is the pressure, $T$ is the temperature, and $V_m$ is the molar volume.
* $a$ is the cohesion pressure parameter (intermolecular attraction).
* $b$ is the co-volume parameter (volume of molecules).

## 1.2 Derivation of Critical Parameters
The critical point $(p_k, V_{mk}, T_k)$ represents a unique state where the distinction between liquid and gas vanishes. Mathematically, the critical isotherm exhibits an inflection point at the critical volume $V_{mk}$. [cite_start]This implies that both the first and second partial derivatives of pressure with respect to volume must vanish[cite: 52]:

$$\left(\frac{\partial p}{\partial V_m}\right)_{T=T_k} = 0 \quad \text{and} \quad \left(\frac{\partial^2 p}{\partial V_m^2}\right)_{T=T_k} = 0$$

[cite_start]Applying these conditions to the van der Waals equation leads to the explicit relations for the critical parameters [cite: 53-57]:
$$V_{mk} = 3b, \quad p_k = \frac{a}{27b^2}, \quad T_k = \frac{8a}{27Rb}$$
This derivation provides the theoretical link between the macroscopic critical point observed in the isotherms and the microscopic van der Waals constants.

## 1.3 The Maxwell Construction
Below the critical temperature ($T < T_k$), the van der Waals equation predicts an oscillating loop which is physically unstable. In reality, the system undergoes a phase transition where liquid and vapor coexist in thermodynamic equilibrium. This requires the chemical potentials of both phases to be equal ($\mu_L = \mu_G$).

In the $p-V$ diagram, this equilibrium is described by the **Maxwell construction**, where the oscillating loop is replaced by a horizontal isotherm (Maxwell line) at the saturation vapour pressure $p_s(T)$. The area between the van der Waals curve and the Maxwell line must vanish ($\oint V dp = 0$).

# 2. Experimental Methodology

## 2.1 Experimental Setup
The isotherms of $SF_6$ were measured using a commercial critical point instrument described in the lab manual [1]. The setup consists of a thick-walled glass capillary containing the substance, which is surrounded by a water bath thermostat to maintain a constant temperature $T$.

The volume of the gas is varied by turning a hand wheel that compresses a column of mercury into the capillary. The pressure $p$ is measured via a manometer connected to the mercury chamber .

## 2.2 Measurement Uncertainties
The measurement uncertainties for volume $u(V)$ and pressure $u(p)$ are estimated based on the reading precision of the analog scales provided in the experimental setup [1].

* **Volume Uncertainty:** The scale on the capillary allows for a reading precision of approx. $\pm 0.05$ ml. Thus, $u(V) = 0.05 \text{ ml}$.
* **Pressure Uncertainty:** The manometer scale allows for a reading precision of approx. $\pm 0.1$ bar. Thus, $u(p) = 0.1 \times 10^5 \text{ Pa}$.

These values are treated as the systematic instrument errors for the subsequent error analysis.

# 4. Task 2: Determination of Saturation Vapour Pressure

## 4.1 Methodology
For temperatures $T < T_c$, the saturation vapour pressure $p_s$ is defined by the pressure of the Maxwell line. We experimentally identified the plateau region for each sub-critical isotherm and calculated $p_s$ as the arithmetic mean of the $N$ pressure readings $p_i$ in this interval:
$$p_s(T) = \frac{1}{N} \sum_{i=1}^N p_i$$

## 4.2 Uncertainty Analysis
The uncertainty $u(p_s)$ combines the systematic instrument error $u(p)_{inst}$ and the statistical fluctuation $\sigma_{stat}$ (standard deviation) of the measurements along the plateau:
$$u(p_s) = \sqrt{\sigma_{stat}^2 + u(p)_{inst}^2}$$
This conservative estimate accounts for both reading limitations and any non-equilibrium fluctuations.

# 5. Task 3: Determination of Average Molar Latent Heat of Vaporization

## 5.1 Theoretical Background
The phase transition between liquid and gas is described by the **Clausius-Clapeyron equation** [1]:
$$\frac{dp_s}{dT} = \frac{Q_{23}}{T \Delta V_m}$$
where $Q_{23}$ is the molar latent heat of vaporization and $\Delta V_m$ is the change in molar volume.

For temperatures significantly below the critical temperature ($T \ll T_c$), two approximations are applied [1]:
1.  The molar latent heat $Q_{23}$ is considered constant.
2.  The vapor phase is treated as an ideal gas ($V_{m,G} \approx RT/p_s$) and the liquid volume is neglected ($V_{m,L} \ll V_{m,G}$).

Integrating the differential equation under these assumptions yields the linearized vapour-pressure equation:
$$\ln(p_s) = \ln(p_{s0}) - \frac{Q_{23}}{R} \cdot \frac{1}{T}$$
This linear relationship allows for the determination of $Q_{23}$ from the slope $m$ of the graph $\ln(p_s)$ vs. $1/T$, where $m = -Q_{23}/R$.

A weighted linear regression was performed on the transformed data $(\ln p_s, 1/T)$ derived from the Task 2 results. The weights were calculated using Gaussian error propagation: $u(\ln p_s) = u(p_s)/p_s$.

# 5. Task 3: Determination of Average Molar Latent Heat of Vaporization

## 5.1 Theoretical Background
The phase transition between liquid and gas is described by the **Clausius-Clapeyron equation** :
$$\frac{dp_s}{dT} = \frac{Q_{23}}{T \Delta V_m}$$
where $Q_{23}$ is the molar latent heat of vaporization and $\Delta V_m$ is the change in molar volume.

For temperatures significantly below the critical temperature ($T \ll T_c$), two approximations are applied:
1.  The molar latent heat $Q_{23}$ is considered constant.
2.  The vapor phase is treated as an ideal gas ($V_{m,G} \approx RT/p_s$) and the liquid volume is neglected ($V_{m,L} \ll V_{m,G}$).

Integrating the differential equation under these assumptions yields the linearized vapour-pressure equation:
$$\ln(p_s) = \ln(p_{s0}) - \frac{Q_{23}}{R} \cdot \frac{1}{T}$$
This linear relationship allows for the determination of $Q_{23}$ from the slope $m$ of the graph $\ln(p_s)$ vs. $1/T$, where $m = -Q_{23}/R$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# --- 1. DATA ENTRY (From your Task 2 Results) ---
T_C = np.array([20.5, 33.0, 36.0, 39.0, 42.0])
ps_bar = np.array([21.03, 28.32, 30.38, 32.71, 35.00])
u_ps_bar = np.array([0.25, 0.22, 0.19, 0.26, 0.33])

# Constants
R = 8.314  # J / (mol K)
T_K = T_C + 273.15

# --- 2. DATA TRANSFORMATION ---
# Model: ln(ps) = ln(p0) - (Q / R) * (1/T)
# x = 1 / T
# y = ln(ps)
x_data = 1 / T_K
y_data = np.log(ps_bar)

# Uncertainty Propagation: u(ln(y)) = u(y) / y
y_err = u_ps_bar / ps_bar

# --- 3. WEIGHTED LINEAR FIT ---
def linear_model(x, m, c):
    return m * x + c

# weighted fit (sigma=y_err)
popt, pcov = curve_fit(linear_model, x_data, y_data, sigma=y_err, absolute_sigma=True)
slope, intercept = popt
slope_err, intercept_err = np.sqrt(np.diag(pcov))

# Calculate Latent Heat Q
# Slope m = -Q / R  =>  Q = -m * R
Q_val = -slope * R
Q_err = slope_err * R

# --- 4. PLOT GENERATION (Publication Style) ---
plt.rcParams.update({
    'font.size': 14,
    'axes.labelsize': 20,
    'xtick.labelsize': 16,
    'ytick.labelsize': 16,
    'legend.fontsize': 14,
    'lines.markersize': 8,
    'lines.linewidth': 2,
    'font.family': 'sans-serif',
    'mathtext.fontset': 'stix',
    'figure.figsize': (10, 7),
    'errorbar.capsize': 4
})

fig, ax = plt.subplots()

# Plot Data
ax.errorbar(x_data, y_data, yerr=y_err, fmt='o', color='black', label='Experimental Data', alpha=0.8)

# Plot Fit
x_fit = np.linspace(min(x_data)*0.99, max(x_data)*1.01, 100)
ax.plot(x_fit, linear_model(x_fit, slope, intercept), 'r-', label='Linear Fit')

# Formatting
# Note: 1/T is usually small, so we often scale it by 10^3 for axis labels,
# but sticking to SI (K^-1) is strictly correct for physics labs.
ax.set_xlabel(r'$1/T$ (K$^{-1}$)')
ax.set_ylabel(r'$\ln(p_s / \text{bar})$')
ax.set_title(r'Vapour-Pressure Curve of $SF_6$')
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend()

plt.tight_layout()
plt.show()

# --- 5. RESULTS OUTPUT ---
print("\n" + "="*60)
print(f"{'TASK 3 RESULTS: Molar Latent Heat':^60}")
print("="*60)
print(f"Fit Slope (m):       {slope:.2f} +/- {slope_err:.2f} K")
print(f"Gas Constant (R):    {R} J/(mol K)")
print("-" * 60)
print(f"Molar Latent Heat Q: {Q_val:.2f} +/- {Q_err:.2f} J/mol")
print(f"                     {Q_val/1000:.2f} +/- {Q_err/1000:.2f} kJ/mol")
print("="*60)
```

# 6. Task 4: Determination of Amount of Substance

## 6.1 Theoretical Background
To determine the amount of substance $n$, we utilize the limiting behavior of real gases. While the van der Waals equation describes the gas at high densities, all gases approach **ideal gas behavior** in the limit of infinite molar volume ($V_m \to \infty$) or zero density ($\rho \to 0$).

The ideal gas law states:
$$pV = nRT$$

For a real gas at a constant temperature $T > T_c$, plotting the product $pV$ against the reciprocal volume $1/V$ yields a curve that intersects the y-axis (where $1/V = 0$) at the value $nRT$. This virial expansion method allows us to determine $n$ without knowing the specific van der Waals constants [1].

## 6.2 Uncertainty Analysis and Error Propagation
A standard linear regression assumes constant uncertainty for all data points. However, in this experiment, the uncertainties vary drastically depending on the volume (heteroscedasticity). We applied Gaussian error propagation to the transformed axes $x = 1/V$ and $y = pV$:

1.  **Uncertainty in $x = 1/V$:**
    $$u(x) = \left| \frac{d}{dV} \left(\frac{1}{V}\right) \right| \cdot u(V) = \frac{u(V)}{V^2}$$
    * **Observation:** The uncertainty scales with $1/V^2$. At **small volumes** (high density, right side of plot), $V^2$ is small, causing the error bars to **explode**. At **large volumes** (low density, left side of plot near the y-axis), $V^2$ is large, causing the error bars to **shrink**.
    * **Significance:** This is advantageous for our analysis. We are extrapolating to the y-axis ($1/V \to 0$). Since the data points closest to the y-axis (large volumes) have the smallest errors, the weighted fit is naturally constrained to be highly precise at the intercept.

2.  **Uncertainty in $y = pV$:**
    $$u(y) = \sqrt{ (p \cdot u(V))^2 + (V \cdot u(p))^2 }$$

To account for these varying uncertainties, a **Weighted Least Squares (WLS)** fit was performed, minimizing $\chi^2 = \sum w_i (y_i - f(x_i))^2$ with weights $w_i = 1/u(y_i)^2$.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# --- 1. DATA PREPARATION ---
# Instrument Uncertainties (from Manual/Setup)
u_V_inst = 0.05  # ml
u_p_inst = 0.1   # bar

# Load Data
try:
    df_raw = pd.read_csv('w15e_readings.csv', header=None)
except FileNotFoundError:
    print("Please upload 'w15e_readings.csv'")
    df_raw = pd.DataFrame()

# Extract T = 50.0 C Data
T_target = 50.0
T_Kelvin = T_target + 273.15
R_gas = 8.314

V_ml = []
p_bar = []

# Robust extraction for T=50 column
for i in range(0, df_raw.shape[1], 2):
    try:
        header = str(df_raw.iloc[0, i])
        if f"T={int(T_target)}" in header or f"T={T_target}" in header:
            data = df_raw.iloc[2:, i:i+2].dropna().astype(float)
            V_ml = data.iloc[:, 0].values
            p_bar = data.iloc[:, 1].values
            break
    except: continue

if len(V_ml) > 0:
    # --- 2. VARIABLE TRANSFORMATION & ERROR PROPAGATION ---
    # Convert to SI Units: V [m^3], p [Pa]
    V_si = V_ml * 1e-6
    p_si = p_bar * 1e5
    u_V_si = u_V_inst * 1e-6
    u_p_si = u_p_inst * 1e5

    # X-axis: x = 1/V
    x_data = 1 / V_si
    # Uncertainty u(x) = |-1/V^2| * u(V)
    x_err = (1 / V_si**2) * u_V_si

    # Y-axis: y = p*V
    y_data = p_si * V_si
    # Uncertainty u(y) = sqrt( (p*uV)^2 + (V*up)^2 )
    y_err = np.sqrt( (p_si * u_V_si)**2 + (V_si * u_p_si)**2 )

    # --- 3. WEIGHTED LINEAR FIT ---
    # We fit y = m*x + c
    # Weights are 1/sigma^2. Scipy uses 'sigma' directly.
    def linear_model(x, m, c):
        return m * x + c

    popt, pcov = curve_fit(linear_model, x_data, y_data, sigma=y_err, absolute_sigma=True)
    slope, intercept = popt
    slope_err, intercept_err = np.sqrt(np.diag(pcov))

    # --- 4. CALCULATE n ---
    # Limit (pV) at 1/V=0 is the intercept 'c'
    # n = c / (RT)
    n_mol = intercept / (R_gas * T_Kelvin)

    # Uncertainty u(n) = u(c) / (RT)
    u_n_mol = intercept_err / (R_gas * T_Kelvin)

    # --- 5. PUBLICATION PLOT ---
    plt.rcParams.update({'font.size': 14, 'axes.labelsize': 20, 'legend.fontsize': 14, 'figure.figsize': (10, 7)})
    fig, ax = plt.subplots()

    # Plot Data with BOTH X and Y Error Bars
    ax.errorbar(x_data, y_data, xerr=x_err, yerr=y_err, fmt='ko',
                label=f'Data ($T={T_target}^\circ$C)', capsize=3, elinewidth=1.5)

    # Plot Fit
    x_fit = np.linspace(0, max(x_data)*1.1, 100)
    ax.plot(x_fit, linear_model(x_fit, slope, intercept), 'r--', label='Weighted Fit')

    # Highlight Intercept
    ax.plot(0, intercept, 'bx', markersize=10, markeredgewidth=2, label='Extrapolation Limit')

    ax.set_xlabel(r'$1/V$ (m$^{-3}$)')
    ax.set_ylabel(r'$pV$ (J)')
    ax.set_title(r'Determination of Amount of Substance $n$')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend(loc='lower right')

    plt.tight_layout()
    plt.show()

    # --- 6. RESULTS OUTPUT ---
    print("\n" + "="*60)
    print(f"{'TASK 4 RESULTS (Rigorous)':^60}")
    print("="*60)
    print(f"Intercept (nRT): {intercept:.4f} +/- {intercept_err:.4f} J")
    print(f"Temperature:     {T_Kelvin:.2f} K")
    print("-" * 60)
    print(f"Amount of Substance n: {n_mol:.5f} +/- {u_n_mol:.5f} mol")
    print(f"                       {n_mol*1000:.2f} +/- {u_n_mol*1000:.2f} mmol")
    print("="*60)
else:
    print("Error: T=50 Data not found.")
```

# 7. Task 5: Temperature Dependence of Latent Heat

## 7.1 Theoretical Derivation
The molar latent heat of vaporization $Q$ (or $\Lambda$) is the energy required to convert one mole of liquid into gas at constant temperature. We calculate it using the **Clausius-Clapeyron equation** with the measured real gas volumes [1]:
$$Q(T) = T \cdot \Delta V_m \cdot \frac{dp_s}{dT}$$

Near the critical temperature $T_c$, the latent heat vanishes according to a power law:
$$Q(T) = Q_0 \left( 1 - \frac{T}{T_c} \right)^\beta$$
where $\beta$ is the critical exponent. This implies that a plot of $Q$ versus the reduced temperature $T/T_c$ should show the curve dropping to zero at $T/T_c = 1$.

## 7.2 Analysis Methodology
1.  **Slope ($dp_s/dT$):** Determined analytically from a polynomial fit to the saturation pressures $p_s(T)$ (Task 2).
2.  **Volume Difference ($\Delta V_m$):** Calculated from the plateau widths $\Delta V_{\text{plateau}}$ measured in Task 1, divided by the amount of substance $n = 1.85 \text{ mmol}$ (Task 4).
3.  **Fit:** The calculated $Q$ values were plotted against the reduced temperature $x = T/T_c$. The data was fitted with the function $Q(x) = A(1-x)^\beta$ to determine the exponent $\beta$.


## 7.3 Uncertainty Analysis
The uncertainty in the latent heat $u(Q)$ is dominated by the reading error of the plateau width, estimated as $u(\Delta V) = \pm 0.05 \text{ ml}$. This error was propagated into the final result:
$$\frac{u(Q)}{Q} \approx \sqrt{ \left( \frac{u(\Delta V)}{\Delta V} \right)^2 + \left( \frac{u(n)}{n} \right)^2 }$$
This results in relative uncertainties that increase significantly as the temperature approaches $T_c$ (where $\Delta V \to 0$), ranging from $\approx 6\%$ at $20.5^\circ\text{C}$ to $\approx 23\%$ at $42^\circ\text{C}$. These large uncertainties are physically inherent to the method and were used to weight the power-law fit.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# --- 1. CONFIGURATION ---
plt.rcParams.update({'font.size': 14, 'axes.labelsize': 20, 'xtick.labelsize': 16,
                     'ytick.labelsize': 16, 'legend.fontsize': 14, 'figure.figsize': (10, 7)})

# Parameters
n_mol = 0.00185
T_c_est = 45.5 + 273.15 # 318.65 K

# Task 2 Data
temps_C = np.array([20.5, 33.0, 36.0, 39.0, 42.0])
temps_K = temps_C + 273.15
ps_bar  = np.array([21.03, 28.32, 30.34, 32.89, 35.25])
ps_Pa   = ps_bar * 1e5

# Measured Widths (dV) and Error
dV_ml = np.array([0.80, 0.45, 0.40, 0.32, 0.22])
u_dV_ml = 0.05

# --- 2. CALCULATIONS ---
# A. Slope dp/dT
poly = np.polyfit(temps_K, ps_Pa, 2)
dp_dT = np.polyval(np.polyder(poly), temps_K)

# B. Molar Volume Difference & Error
dV_si = dV_ml * 1e-6
dV_m = dV_si / n_mol
# Error propagation for dV_m
u_n_mol = 0.00001
u_dV_m = dV_m * np.sqrt( (u_dV_ml/dV_ml)**2 + (u_n_mol/n_mol)**2 )

# C. Latent Heat Q & Error
Q_vals = temps_K * dV_m * dp_dT
u_Q_vals = Q_vals * (u_dV_m / dV_m) # Dominant error is dV

# --- 3. FIT VS REDUCED TEMP (T / Tc) ---
# X-axis: T / Tc (Values like 0.92, 0.95...)
x_data = temps_K / T_c_est

# Model: Q(x) = A * (1 - x)^beta
def power_law_standard(x, A, beta):
    return A * np.power(1 - x, beta)

# Fit
popt, pcov = curve_fit(power_law_standard, x_data, Q_vals, sigma=u_Q_vals, absolute_sigma=True, p0=[10000, 0.35])
A_fit, beta_fit = popt
A_err, beta_err = np.sqrt(np.diag(pcov))

# --- 4. PLOT ---
fig, ax = plt.subplots()

# Plot Data
ax.errorbar(x_data, Q_vals, yerr=u_Q_vals, fmt='ko', capsize=4, label='Experimental Data')

# Plot Fit Curve
x_fit = np.linspace(min(x_data)*0.99, 0.999, 100) # Plot up to just before 1.0
y_fit = power_law_standard(x_fit, *popt)
ax.plot(x_fit, y_fit, 'r-', linewidth=2, label=f'Fit: $\\beta = {beta_fit:.2f} \pm {beta_err:.2f}$')

# Formatting
ax.set_xlabel(r'Reduced Temperature $T / T_c$')
ax.set_ylabel(r'Latent Heat $Q$ (J/mol)')
ax.set_title(r'Temperature Dependence of Latent Heat')
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend()

plt.tight_layout()
plt.show()

# --- 5. RESULTS ---
print("\n" + "="*80)
print(f"{'TASK 5 RESULTS (Strict T/Tc Plot)':^80}")
print("="*80)
print(f"Critical Exponent beta: {beta_fit:.3f} +/- {beta_err:.3f}")
print("="*80)
```
